```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import scipy.stats as stats
```

```python
poisson_means = lambda x: [np.mean(np.random.poisson(size=(x,))) for _ in range(10000)]
uniform_means = lambda x: [np.mean(np.random.uniform(size=(x,))) for _ in range(10000)]
exponential_means = lambda x: [np.mean(np.random.exponential(size=(x,))) for _ in range(10000)]
```

```python
def plot(data):
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))
    ax[0].hist(data, histtype='step');
    stats.probplot(data, plot=ax[1]);
```

# Poisson

```python
plot(poisson_means(10))
```

```python
plot(poisson_means(100))
```

```python
plot(poisson_means(1000))
```

# Uniform

```python
plot(uniform_means(10))
```

```python
plot(uniform_means(100))
```

# Exponential

```python
plot(exponential_means(10))
```

```python
plot(exponential_means(100))
```

```python
plot(exponential_means(1000))
```

```python
plot(exponential_means(10000))
```
