```python
import numpy as np
import matplotlib.pyplot as plt
```

# Mock gaussian log likelihood

```python
def mock_gaussian_log_likelihood(x, stddev=1., mean=5.):
    return np.exp(-0.5*np.power(np.subtract(x, mean), 2)/
                  (np.power(stddev, 2)))/(stddev*np.sqrt(2*np.pi))
```

```python
x = np.linspace(0, 10, 1000)
log_likelihood = mock_gaussian_log_likelihood(x)
plt.plot(x, log_likelihood)
plt.grid(linestyle='--')
```

```python
def integrate_for_likelihood(val, lvalues):
    return np.sum([v for v in lvalues
                   if v>=val])/np.sum(lvalues)
```

```python
val_at_sigma = log_likelihood[np.argmin(np.abs(x-4.))]
integrate_for_likelihood(val_at_sigma, log_likelihood)
```

```python
def binsearch(low, high, target, tolerance, i=0, max_i=10):
    middle = (low+high)/2
    confidence = integrate_for_likelihood(middle, log_likelihood)
    print(f'Iteration {i}: {middle}. Confidence: {confidence}')
    
    if i>max_i:
        print(f'Maximum number of iterations ({max_i}) exceeded')
        return middle
    
    if low>high:
        print(f'low {low} > high {high}')
        return middle
    
    if np.isclose(confidence, target, rtol=tolerance):
        return middle
    
    if confidence<=target:
        return binsearch(low, middle, target, tolerance, i+1, max_i)
    else:
        return binsearch(middle, high, target, tolerance, i+1, max_i)
```

```python
conf = binsearch(0.2, 0.3, 0.683, 1e-3)
```

```python
plt.plot(x, log_likelihood)
plt.hlines(conf, 0, 10, color='red')
plt.grid(linestyle='--')
```

# Composite Gaussian

```python
def multigaussian(x):
    return 1.5*mock_gaussian_log_likelihood(x, mean=1., stddev=3.)+0.8*mock_gaussian_log_likelihood(x, mean=8., stddev=1.5)
```

```python
x = np.linspace(0, 10, 1000)
log_likelihood = multigaussian(x)
plt.plot(x, log_likelihood)
plt.grid(linestyle='--')
```

```python
integrate_for_likelihood(0.16, log_likelihood)
```

```python
integrate_for_likelihood(0.14, log_likelihood)
```

```python
conf = binsearch(0.14, 0.16, 0.683, 1e-3)
```

```python
plt.plot(x, log_likelihood)
plt.hlines(conf, 0, 10, color='red')
plt.grid(linestyle='--')
```
