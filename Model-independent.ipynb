```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([7., 10., 11., 10., 8., 14., 12., 9., 12., 13.,
        10., 11., 12., 11., 13., 9., 10., 12., 11.,
        7., 7., 12., 10., 6., 8., 11., 9., 9., 9., 9.])
```

```python
errors = np.array([3.18, 3.45, 2.92, 3.14, 3.43, 2.85, 3.18,
         3.14, 3.12, 3.03, 3.12, 3.07, 3.08, 3.41,
         3.32, 3.44, 3.31, 3.21, 2.87, 3.07, 2.89,
         3.06, 2.93, 2.90, 3.32, 2.99, 3.08, 3.18,
         3.40, 3.01])
```

```python
def gaussian_log_likelihood(data, errors, params):
    g_mean = params[0]
    return np.longdouble(-0.5*np.sum(np.power((data-g_mean)/errors, 2)))
```

```python
means = np.linspace(0, 20, 100)
likelihoods = np.exp([gaussian_log_likelihood(data, errors, [p]) for p in means])
```

```python
gaussian_log_likelihood(data, errors, [5.])
```

```python
plt.plot(means, likelihoods)
plt.grid(linestyle='--')
```

## Resampling

For each datapoint $x_n$, invoke a Gaussian with mean $x_n$ and standard deviation $\sigma_n$ as measured during the experiment. We can now sample a new data point $x'_n$ from this distribution.

```python
np.random.normal(loc=data[0], scale=errors[0])
```

```python
def resample(data, errors):
    return np.array([np.random.normal(
        loc=d, scale=e
    ) for d, e in zip(data, errors)])
```

```python
x = list(range(len(data)))
plt.errorbar(x=x, y=data, yerr=errors)
plt.errorbar(x=x, y=resample(data, errors), yerr=errors, color='red')
plt.grid(linestyle='--')
```

```python
resampled_means = [np.mean(resample(data, errors)) for _ in range(100000)]
```

```python
counts, steps, _ = plt.hist(resampled_means, bins=50);
plt.grid(linestyle='--')
```

```python
def normal_stddev(means, counts):
    m = means.dot(counts)
    mom2 = np.power(means, 2).dot(counts)
    variance = mom2 - np.power(m, 2)
    return np.sqrt(variance)
```

```python
np.mean(resampled_means)
```

```python
normal_stddev(steps[:-1], counts/np.sum(counts))
```

## Bootstrapping

Resample data to create alternative datasets. We draw new samples from the measured dataset itself.

```python
bootstrap_means = [np.mean(np.random.choice(data, size=(len(data),)))
                   for _ in range(100000)]
```

```python
counts, steps, _ = plt.hist(bootstrap_means, bins=90);
plt.grid(linestyle='--')
```

```python
np.mean(bootstrap_means)
```

```python
normal_stddev(steps[:-1], counts/np.sum(counts))
```
